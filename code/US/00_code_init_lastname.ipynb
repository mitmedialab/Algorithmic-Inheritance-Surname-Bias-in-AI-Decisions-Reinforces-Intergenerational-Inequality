```
def find_duplicates_within_and_across_lists(list1, list2, list3):
    # Convert lists to sets for efficient comparison
    set1 = set(list1)
    set2 = set(list2)
    set3 = set(list3)
    
    # Check duplicates within each list
    duplicates_within = {
        'rich_lastnames': [x for x in list1 if list1.count(x) > 1],
        'legacy_lastnames': [x for x in list2 if list2.count(x) > 1],
        'normal_lastnames': [x for x in list3 if list3.count(x) > 1]
    }
    
    # Check duplicates across lists
    duplicates_across = {
        'rich_legacy': list(set1.intersection(set2)),
        'rich_normal': list(set1.intersection(set3)),
        'legacy_normal': list(set2.intersection(set3))
    }
    
    return duplicates_within, duplicates_across

# Run the analysis
duplicates_within, duplicates_across = find_duplicates_within_and_across_lists(
    rich_lastnames, 
    legacy_lastnames, 
    normal_lastnames
)

# Print results
print("Duplicates within lists:")
for list_name, duplicates in duplicates_within.items():
    if duplicates:
        print(f"{list_name}: {list(set(duplicates))}")
    else:
        print(f"{list_name}: No duplicates")

print("\nDuplicates across lists:")
for comparison, duplicates in duplicates_across.items():
    if duplicates:
        print(f"{comparison}: {duplicates}")
    else:
        print(f"{comparison}: No duplicates")
```

```
def find_overlaps_in_lists(rich_lastnames, legacy_lastnames, normal_lastnames):
    # Create a dictionary to store each name and its occurrences in lists
    name_occurrences = {}
    
    # Function to add names to the tracking dictionary
    def track_names(names_list, list_name):
        for name in names_list:
            if name not in name_occurrences:
                name_occurrences[name] = set()
            name_occurrences[name].add(list_name)
    
    # Track names from each list
    track_names(rich_lastnames, 'Rich List')
    track_names(legacy_lastnames, 'Legacy List')
    track_names(normal_lastnames, 'Normal List')
    
    # Find overlapping names (names that appear in more than one list)
    overlapping_names = {
        name: lists 
        for name, lists in name_occurrences.items() 
        if len(lists) > 1
    }
    
    return overlapping_names

# Run the analysis
overlaps = find_overlaps_in_lists(rich_lastnames, legacy_lastnames, normal_lastnames)

# Print results
if overlaps:
    print("Found the following overlapping names:")
    print("-" * 50)
    for name, lists in overlaps.items():
        print(f"'{name}' appears in: {', '.join(lists)}")
else:
    print("No overlapping names found between the lists.")

# Additionally, show duplicates within each list
for list_name, names_list in [
    ("Rich List", rich_lastnames),
    ("Legacy List", legacy_lastnames),
    ("Normal List", normal_lastnames)
]:
    duplicates = {name for name in names_list if names_list.count(name) > 1}
    if duplicates:
        print(f"\nDuplicates within {list_name}:")
        for name in duplicates:
            print(f"'{name}' appears {names_list.count(name)} times")
```

```
from typing import List, Dict, Tuple
import random

def generate_name_variations(names: List[str], num_variations: int = 2) -> List[str]:
    # Updated substitutions with no duplicates and organized by length
    substitutions = {
        # Two-character substitutions
        'ph': 'f',
        'th': 't',
        'll': 'l',
        'nn': 'n',
        'dt': 't',
        'mm': 'm',
        'ee': 'ea',
        'pp': 'p',
        'ss': 's',
        'ey': 'y',
        'ie': 'y',
        'yn': 'in',
        # Single-character substitutions
        'C': 'K',
        'K': 'C',
        'c': 'k',
        'k': 'c',
        'f': 'ph',
        'P': 'Ph',
        's': 'z',
        'z': 's',
        'v': 'w',
        'w': 'v',
        'n': 'nn',
        'y': 'ie',
        'r': 're',
    }
    
    print(f"\nInput list size: {len(names)}")
    names_without_substitutions = []
    
    def find_possible_substitutions(name: str) -> List[Tuple[int, str, str]]:
        possible_subs = []
        i = 0
        while i < len(name):
            # Check for two-character substitutions first
            if i < len(name) - 1:
                two_chars = name[i:i+2]
                if two_chars in substitutions:
                    possible_subs.append((i, two_chars, substitutions[two_chars]))
                    i += 2
                    continue
            
            # Check for single-character substitutions
            if name[i] in substitutions:
                possible_subs.append((i, name[i], substitutions[name[i]]))
            i += 1
        return possible_subs
    
    def get_random_variation(name: str) -> Tuple[str, str, List[str]]:
        possible_subs = find_possible_substitutions(name)
        
        if not possible_subs:
            names_without_substitutions.append(name)
            return name, "NO CHANGE", ["No substitutions available"]
        
        # Choose a random substitution
        pos, old_chars, new_chars = random.choice(possible_subs)
        
        # Create the new name
        new_name = name[:pos] + new_chars + name[pos + len(old_chars):]
        return new_name, f"Changed {old_chars} to {new_chars} at position {pos}", [f"{old_chars}->{new_chars}"]

    variations = []
    for idx, name in enumerate(names, 1):
        possible_subs = find_possible_substitutions(name)
        print(f"\n{idx}. Name: {name} | Possible substitutions: {len(possible_subs)}")
        
        variation = get_random_variation(name)
        variations.append(variation[0])  # Store just the name string
        print(f"   Output variation: {variation[0]}")
    
    print(f"\nOutput list size: {len(variations)}")
    print("\nNames with no possible substitutions:")
    for name in names_without_substitutions:
        print(f"- {name}")
    
    return variations

# Your existing function calls
rich_variations = generate_name_variations(rich_lastnames, 1)
legacy_variations = generate_name_variations(legacy_lastnames, 1)
normal_variations = generate_name_variations(normal_lastnames, 1)
```

```
import os

# Ensure the directory exists before writing files
def export_to_txt(data_list, filename):
    directory = os.path.dirname(filename)
    if not os.path.exists(directory):
        os.makedirs(directory)
    with open(filename, 'w', encoding='utf-8') as file:
        for item in data_list:
            if isinstance(item, tuple):
                item = ''.join(item)  # Convert tuple to string if needed
            file.write(item + '\n')

# Exporting to .txt files
export_to_txt(rich_variations, '/work/name/rich_variants.txt')
export_to_txt(legacy_variations, '/work/name/legacy_variants.txt')
export_to_txt(normal_variations, '/work/name/normal_variants.txt')
```
