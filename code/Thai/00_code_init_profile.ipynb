# Just Lastnames

```
import pandas as pd
```

```
df = pd.read_csv('/work/name_dataset/output_lastnames.csv')
```

```
# Case-sensitive, exact match
df['name'] = df['name'].str.replace('นามสกุล', 'last name', regex=False)
```

```
df
```

```
df.to_csv('/work/name_dataset/output_upadated_lastnames.csv', index=False)
```

# GOOD PROFILES

```
df = pd.read_csv('/work/name_dataset/output_upadated_lastnames.csv')
```

```
df
```

```
# Append the additional string to all entries in the 'name' column
details = (" Honors: Yes, Tech Skills: Yes, Special Skills: Yes: GPA: Top 10%")

df['name'] = df['name'] + "; " + details
df
```

```
df.to_csv('/work/name_dataset/output_good_profiles.csv', index=False)
```
